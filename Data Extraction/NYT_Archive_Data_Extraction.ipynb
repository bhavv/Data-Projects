```python
#As headline column consists of nested dicts, we extract the data contents in headline column into a new df with _id
```

```python
import ast

# Copying the DataFrame
df = archive_df.copy()

#Extracting the keys and values from 'headline' column
headlines_data = []
for idx, row in df.iterrows():
    #Converting the string representation of dictionary to a dictionary
    row_dict = ast.literal_eval(row['headline'])
    #Adding _id to the dictionary
    row_dict['_id'] = row['_id']
    headlines_data.append(row_dict)

#Creating a DataFrame from the extracted data
headlines_df = pd.DataFrame(headlines_data)

#Reordering columns to have '_id' as the first column
columns = ['_id'] + [col for col in headlines_df.columns if col != '_id']
headlines_df = headlines_df[columns]

#Saving DataFrame to a CSV file
#headlines_df.to_csv(r'C:\HOME\SJSU\DBMS\NYT\archive data\Extracted\headlines_data.csv', index=False)
```

```python
#extracted headline
headlines_df.head(3)
```

```python
headlines_df.rename(columns={'name': 'headlines_name'}, inplace=True)
```

```python
#As keywords column consists of list of nested dicts, we extract the data contents in keywords column into a new df with _id
```

```python
#extracting keywords column 
keywords_data = []

#Iterating each row in the df
for index, row in archive_df.iterrows():
    #Extracting the _id and keywords
    _id = row["_id"]
    keywords = row["keywords"]
    
    #Checking if the keywords column is not null
    if pd.notnull(keywords):
        #Converting string to list of dictionaries
        keywords_list = eval(keywords)
        
        #Iterating over each keyword in the keywords list
        for keywords in keywords_list:
            #Appending the keyword data along with the _id to the keywords_data list
            keywords_data.append({
                "_id": _id,
                "name": keywords['name'],
                "value": keywords['value'],
                "rank": keywords['rank'],
                "major" : keywords['major']
            })

#Creating DataFrame from the list
keywords_df = pd.DataFrame(keywords_data)

#Saving DataFrame to a CSV file
#keywords_df.to_csv(r'C:\HOME\SJSU\DBMS\NYT\archive data\Extracted\keywords_data.csv', index=False)
```

```python
#extracted keywords
keywords_df.head(3)
```

```python
keywords_df.rename(columns={'name': 'keyword_name'}, inplace=True)
```

```python
#As byline consists of list of nested dicts, we extract the data contents in byline column into a new df with _id
```

```python
df = archive_df.copy()

#Extracting the keys and values from 'byline' column
byline_data = []
for idx, row in df.iterrows():
    #Converting the string representation of dictionary to a dictionary
    row_dict = ast.literal_eval(row['byline'])
    #Adding _id to the dictionary
    row_dict['_id'] = row['_id']
    byline_data.append(row_dict)

#Creating a DataFrame from the extracted data
byline_df = pd.DataFrame(byline_data)

#Reordering columns to have '_id' as the first column
columns = ['_id'] + [col for col in byline_df.columns if col != '_id']
byline_df = byline_df[columns]

#Saving DataFrame to a CSV file
#byline_df.to_csv(r'C:\HOME\SJSU\DBMS\NYT\archive data\Extracted\byline_data.csv', index=False)
```
