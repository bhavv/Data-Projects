```python
import pandas as pd
```

```python
#keyword csv
dataset = pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\keywords.csv")

keywords_data = []

for index, row in dataset.iterrows():
    movie_id = row["id"]
    keywords = row["keywords"]
    
    if pd.notnull(keywords):
        keywords_list = eval(keywords)
        
        for keyword in keywords_list:
            keywords_data.append({
                "movie_id": movie_id,
                "keywords_ID": keyword['id'],
                "keywords_name": keyword['name']                
            })

#Creating DataFrame from the list
keywords_df = pd.DataFrame(keywords_data)

#Writing DataFrame to a CSV file
keywords_df.to_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\keywords_data.csv", index=False)

print("DONE")
```

```python
#cast and crew
import pandas as pd

credits = pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\credits.csv")
cast_data = []
crew_data = []

for index, row in credits.iterrows():
    id2 = row["id"]
    cast = row["cast"]
    crew = row["crew"]
    
    #Processing the cast information
    if pd.notnull(cast):
        cast_list = eval(cast)
        for actor in cast_list:
            cast_data.append({
                "Id": id2,
                "Cast_Character_ID": actor.get('cast_id'),
                "Cast_Character": actor.get('character'),
                "Cast_Credit_ID": actor.get('credit_id'),
                "Cast_gender": actor.get('gender'),
                "Cast_id": actor.get('id'),
                "Cast_name": actor.get('name'),
                "Cast_order": actor.get('order'),
                "Cast_profile_path": actor.get('profile_path')
            })
    
    #Processing the crew information
    if pd.notnull(crew):
        crew_list = eval(crew)
        for member in crew_list:
            crew_data.append({
                "Id": id2,
                "Crew_Credit_ID": member.get('credit_id'),
                "Crew_Department": member.get('department'),
                "Crew_gender": member.get('gender'),
                "Crew_Job": member.get('job'),
                "Crew_Id": member.get('id'),
                "Crew_name": member.get('name'),
                "Crew_profile_path": member.get('profile_path')
            })

#Creating DataFrames from the lists
cast_df = pd.DataFrame(cast_data)
crew_df = pd.DataFrame(crew_data)

#Writing DataFrames to CSV files
cast_df.to_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\cast_data.csv", index=False)
crew_df.to_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\crew_data.csv", index=False)

print("DONE")
```

```python
#meta data separates
import pandas as pd

movies_metadata = pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\movies_metadata.csv", dtype={10: str})

# Initialize empty lists to store extracted data
Movies_belongs_to_collection_data = []
Movies_genres_data = []
Movies_production_companies_data = []
Movies_production_countries_data = []
Movies_spoken_languages_data = []

# Iterate over each row in the DataFrame
for index, row in movies_metadata.iterrows():
    id3 = row["id"]
    # Process belongs_to_collection
    belongs_to_collection = row["belongs_to_collection"]
    if pd.notnull(belongs_to_collection):
        try:
            collection_dict = eval(belongs_to_collection)
            Movies_belongs_to_collection_data.append({
                "ID": id3,
                "belongs_to_collection_id": collection_dict.get('id'),
                "belongs_to_collection_name": collection_dict.get('name'),
                "belongs_to_collection_poster_path": collection_dict.get('poster_path'),
                "belongs_to_collection_backdrop_path": collection_dict.get('backdrop_path')
            })
        except Exception as e:
            print(f"Error processing belongs_to_collection: {e}")
    
    # Process genres
    genres = row["genres"]
    if pd.notnull(genres):
        try:
            genres_list = eval(genres)
            for genre in genres_list:
                Movies_genres_data.append({
                    "ID": id3,
                    "Movies_genres_ID": genre.get('id'),
                    "Movies_genres_name": genre.get('name')
                })
        except Exception as e:
            print(f"Error processing genres: {e}")
    
    # Process production_companies
    production_companies = row["production_companies"]
    if pd.notnull(production_companies):
        try:
            companies_list = eval(production_companies)
            for company in companies_list:
                Movies_production_companies_data.append({
                    "ID": id3,
                    "Movies_production_companies_name": company.get('name'),
                    "Movies_production_companies_id": company.get('id')
                })
        except Exception as e:
            print(f"Error processing production_companies: {e}")
    
    # Process production_countries
    production_countries = row["production_countries"]
    if pd.notnull(production_countries):
        try:
            countries_list = eval(production_countries)
            for country in countries_list:
                Movies_production_countries_data.append({
                    "ID": id3,
                    "Movies_production_countries_iso_3166_1": country.get('iso_3166_1'),
                    "Movies_production_countries_name": country.get('name')
                })
        except Exception as e:
            print(f"Error processing production_countries: {e}")
    
    # Process spoken_languages
    spoken_languages = row["spoken_languages"]
    if pd.notnull(spoken_languages):
        try:
            languages_list = eval(spoken_languages)
            for language in languages_list:
                Movies_spoken_languages_data.append({
                    "ID": id3,
                    "Movies_spoken_languages_iso_639_1": language.get('iso_639_1'),
                    "Movies_spoken_languages_name": language.get('name')
                })
        except Exception as e:
            print(f"Error processing spoken_languages: {e}")

# Convert the lists to DataFrames
Movies_belongs_to_collection_df = pd.DataFrame(Movies_belongs_to_collection_data)
Movies_genres_df = pd.DataFrame(Movies_genres_data)
Movies_production_companies_df = pd.DataFrame(Movies_production_companies_data)
Movies_production_countries_df = pd.DataFrame(Movies_production_countries_data)
Movies_spoken_languages_df = pd.DataFrame(Movies_spoken_languages_data)

# Write DataFrames to CSV files
Movies_belongs_to_collection_df.to_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Movies_belongs_to_collection_data.csv", index=False)
Movies_genres_df.to_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Movies_genres_data.csv", index=False)
Movies_production_companies_df.to_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Movies_production_companies_data.csv", index=False)
Movies_production_countries_df.to_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Movies_production_countries_data.csv", index=False)
Movies_spoken_languages_df.to_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Movies_spoken_languages_data.csv", index=False)

print("Extraction and CSV writing completed successfully.")
```
