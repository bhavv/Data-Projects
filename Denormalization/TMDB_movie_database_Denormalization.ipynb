```python
#bhmv:I have constructed denormalization code. Please have a look! This code denormalizes the data and pushes it to the mongo cluster.
import pandas as pd
from pymongo import MongoClient
import json
```

```python
#loading cleaned data as Dataframes
Movies_meta_data= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\movies_metadata.csv", low_memory=False)
Movies_belongs_to_collection= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\Movies_belongs_to_collection_data.csv")
Movies_genres_df= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\Movies_genres_data.csv")
Movies_production_companies_df= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\Movies_production_companies_data.csv")
Movies_production_countries_df= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\Movies_production_countries_data.csv")
Movies_spoken_languages_df= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\Movies_spoken_languages_data.csv")
cast= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\cast_data.csv")
crew= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\crew_data.csv")
keywords= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\keywords_data.csv")
ratings= pd.read_csv(r"C:\HOME\SJSU\DBMS\LAB\Downloads\Mongo\ratings_data.csv")
```

```python
#Merging Movies_meta_data and Movies_belongs_to_collection as merged_df
```

```python
collections_array_df = Movies_belongs_to_collection.groupby('Id')['belongs_to_collection_name'].apply(list).reset_index(name='collections')
```

```python
merged_df = Movies_meta_data.merge(collections_array_df, on='Id', how='left')
```

```python
#Merging merged_df and keywords
```

```python
k_array_df = keywords.groupby('Id')['keywords_name'].apply(list).reset_index(name='keywords')
```

```python
merged_df = merged_df.merge(k_array_df, on='Id', how='left')
```

```python
#Merging merged_df and Movies_genres_df
```

```python
genres_array_df = Movies_genres_df.groupby('Id')['Movies_genres_name'].apply(list).reset_index(name='genres')
```

```python
merged_df = merged_df.merge(genres_array_df, on='Id', how='left')
```

```python
#Merging merged_df and Movies_production_companies_df
```

```python
pr_array_df = Movies_production_companies_df.groupby('Id')['Movies_production_companies_name'].apply(list).reset_index(name='production_companies')
```

```python
merged_df = merged_df.merge(pr_array_df, on='Id', how='left')
```

```python
#Merging merged_df and Movies_production_countries_df
```

```python
pc_array_df = Movies_production_countries_df.groupby('Id')['Movies_production_countries_name'].apply(list).reset_index(name='production_countries')
```

```python
merged_df = merged_df.merge(pc_array_df, on='Id', how='left')
```

```python
#Merging merged_df and Movies_spoken_languages_df
```

```python
s_array_df = Movies_spoken_languages_df.groupby('Id')['Movies_spoken_languages_name'].apply(list).reset_index(name='spoken_languages')
```

```python
merged_df = merged_df.merge(s_array_df, on='Id', how='left')
```

```python
#Merging merged_df and cast
```

```python
merged_df = merged_df.merge(cast.groupby('Id').apply(lambda x: x.drop(columns='Id').to_dict('records')).reset_index(name='cast_details'), on='Id', how='left')
```

```python
#Merging merged_df and crew
```

```python
merged_df = merged_df.merge(crew.groupby('Id').apply(lambda x: x.drop(columns='Id').to_dict('records')).reset_index(name='crew_details'), on='Id', how='left')
```

```python
#Merging merged_df and ratings
```

```python
merged_df = merged_df.merge(ratings.groupby('Id').apply(lambda x: x.drop(columns='Id').to_dict('records')).reset_index(name='ratings_data'), on='Id', how='left')
```
